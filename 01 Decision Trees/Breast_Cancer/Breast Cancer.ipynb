```python
#Importamos librerias
import numpy as np
import pandas as pd
```

```python
#Importamos nuestro .csv (data set)
cancer = pd.read_csv('wisc_bc_data.csv')
cancer.head(n=80)
```

```python
#cancer = cancer.iloc[:, 2:]
#cancer.head()
```

```python
cancer['diagnosis'].value_counts(), cancer['diagnosis'].value_counts(normalize = True)
```

```python
#93%#X = cancer[['symmetry_worst','texture_mean','smoothness_mean', 'concavity_mean','compactness_mean', 'radius_mean', 'area_mean', 'perimeter_mean','texture_worst', 'radius_mean']]
X = cancer[['radius_mean', 'radius_worst', 'points_mean', 'area_mean', 'area_worst', 'perimeter_mean', 'perimeter_worst', 'concavity_mean', 'symmetry_worst', 'dimension_worst', 'compactness_mean']]
X.head()
```

```python
y = cancer.diagnosis
y.head()
```

```python
from sklearn.model_selection import train_test_split # solito hace un muestreo aleatorio

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 123)
```

```python
len(cancer)
```

```python
len(y_train), len(y_test)
```

```python
from sklearn import tree
```

```python
tree_clf = tree.DecisionTreeClassifier(max_depth = 3)
```

```python
tree_clf.fit(x_train, y_train)
```

```python
y_pred = tree_clf.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

accuracy = accuracy_score(y_test, y_pred)
print('DecisionTreeClassifier accuracy score: {}'.format(accuracy))
```
